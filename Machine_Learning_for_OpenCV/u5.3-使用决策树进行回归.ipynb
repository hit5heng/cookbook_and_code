# 使用决策树进行回归 
tree.DecisionTreeRegressor

```python
import numpy as np
rng = np.random.RandomState(42)

# 创建100个0-5的点
x = np.sort(5 * rng.rand(100, 1), axis=0)
y = np.sin(x).ravel()
```

```python
# 添加噪声
y[::2] += 0.5 * (0.5 - rng.rand(50))
```

## 设置不同深度的树，使用默认的‘mse’

```python
from sklearn import tree
reg1 = tree.DecisionTreeRegressor(max_depth=2, random_state=42)
reg1.fit(x, y)

reg2 = tree.DecisionTreeRegressor(max_depth=5, random_state=42)
reg2.fit(x, y)
```

np.arange
- 1）一个参数时，参数值为终点，起点取默认值0，步长取默认值1。 
- 2）两个参数时，第一个参数为起点，第二个参数为终点，步长取默认值1。 
- 3）三个参数时，第一个参数为起点，第二个参数为终点，第三个参数为步长。其中步长支持小数。

```python
x_test = np.arange(0.0, 5.0, 0.01)[:, np.newaxis]

y_1 = reg1.predict(x_test)
y_2 = reg2.predict(x_test)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.style.use('ggplot')
plt.figure(figsize=(16, 12))

plt.scatter(x, y, c='k', s=50, label='data')
plt.plot(x_test, y_1, label='max_depth=2', linewidth=5)
plt.plot(x_test, y_2, label='max_depth=5', linewidth=5)
plt.xlabel('data')
plt.ylabel('target')
plt.legend()
plt.show()
```
