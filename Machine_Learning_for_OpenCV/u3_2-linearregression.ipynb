# 线性回归 
- linear_model.LinearRegression()


```python
from sklearn import datasets, metrics, linear_model
from sklearn import model_selection as modsel
import matplotlib.pyplot as plt

plt.style.use('ggplot')
%matplotlib inline
```

## 数据集：波士顿房价

```python
boston = datasets.load_boston()
print(dir(boston))  # ['DESCR', 'data', 'feature_names', 'filename', 'target']
print('data shape', boston.data.shape)  # data shape (506, 13)
print('target shape', boston.target.shape)  # target shape (506,)
```

## 回归算法选择
- 线性回归： linear_model.LinearRegression()
- L1正则化： L1范数（也叫直线距离、蛇形距离、曼哈顿距离）
    - Lasso回归算法： linear_model.Lasso()
- L2正则化： L2范数（也叫欧式距离），对离群值惩罚力度比L1大
    - ridge回归算法： linear_model.Ridge()

```python
linreg = linear_model.LinearRegression()
# linreg = linear_model.Lasso()  # Lasso 回归算法（L1）
# linreg = linear_model.Ridge()  # ridge 回归算法（L2）
```

### 数据分割

```python
x_train, x_test, y_train, y_test = modsel.train_test_split(boston.data, boston.target, test_size=0.1, random_state=42)
# print(x_train.shape)  # (455, 13)
# print(y_train.shape)  # (455,)
# print(x_test.shape)  # (51, 13)
# print(y_test.shape)  # (51,)
```

### train

```python
# sklearn中的train函数：fit
linreg.fit(x_train, y_train)
```

```python
print(metrics.mean_squared_error(y_train, linreg.predict(x_train)))  # 预测值的均方误差（mse）
print(linreg.score(x_train, y_train))  # 返回确定系数（R方值）

# equation coefficient and intercept
print("系数", linreg.coef_)  # 打印系数
print("截距", linreg.intercept_)  # 打印截距
```

### test

```python
y_pred = linreg.predict(x_test)
print(metrics.mean_squared_error(y_pred, y_test))
```

```python
# 显示回归图形
plt.figure(figsize=(10, 6))
plt.plot(y_test, linewidth=3, label='truth')
plt.plot(y_pred, linewidth=3, label='predicted')
plt.legend(loc='best')
plt.xlabel('test data points')
plt.ylabel('target value')
plt.show()
```
