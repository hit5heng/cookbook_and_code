# K均值一个有趣的应用——压缩图像的色彩空间

```python
import cv2
import numpy as np

lena = cv2.imread('data/lena.jpg', cv2.IMREAD_COLOR)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use("ggplot")

# 禁用ggplot的网格线
plt.rc('axes', **{'grid': False})
plt.imshow(cv2.cvtColor(lena, cv2.COLOR_BGR2RGB))
```

```python
lena.shape
```

```python
img_data = lena/255.0  # 归一化
img_data = img_data.reshape((-1,3))  # -1 自动选择
img_data.shape
```

```python
def plot_pixels(data, title, colors=None, N=10000):
    if colors is None:
        colors = data
    
    rng = np.random.RandomState(0)
    i = rng.permutation(data.shape[0])[:N]
    colors = colors[i]
    
    pixel = data[i].T
    R, G, B = pixel[0], pixel[1], pixel[2]
    
    fig, ax = plt.subplots(1, 2, figsize=(16,6))
    ax[0].scatter(R, G, color=colors, marker='.')
    ax[0].set(xlabel='RED', ylabel='GREEN', xlim=(0,1), ylim=(0,1))
    
    ax[1].scatter(R, B, color=colors, marker='.')
    ax[1].set(xlabel='RED', ylabel='BLUE', xlim=(0,1), ylim=(0,1))
    
    fig.suptitle(title, size=20)
```

```python
plot_pixels(img_data,title='Input color space: 16 million possible color')
```

## 使用K均值减少调色板

```python
# 指定参数：迭代10次，误差小于1.0
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)
flags = cv2.KMEANS_RANDOM_CENTERS

img_data = img_data.astype(np.float32)

compactness, labels, centers = cv2.kmeans(img_data,
                                         16,
                                         None,
                                         criteria,
                                         10,
                                         flags)
```

```python
centers
```

```python
new_colors = centers[labels].reshape((-1,3))
```

```python
new_colors.shape
```

```python
plot_pixels(img_data, colors=new_colors, title='Reduce color space: 16 colors')
```

```python
lena_recolored = new_colors.reshape(lena.shape)
```

```python
plt.imshow(cv2.cvtColor(lena_recolored, cv2.COLOR_BGR2RGB))
plt.title('16-color image')
```
