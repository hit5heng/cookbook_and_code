面试17题：

题目：打印从1到最大的n位数

题：输入数字n,按顺序打印出从1到最大的n位十进制数，比如输入3，则打印出1、2、3一直到最大的3位数999.

 

解题思路：需要考虑大数问题，这是题目设置的陷阱。可以把问题转换成数字排列问题，用递归让代码更简洁。 参见剑指offer P114

```python
class Solution:
    
    def print1ToMaxOfNDigits(self, n):
        if n <= 0:
            return
        
        number = ["0"] * n
        for i in range(10):
            number[0] = str(i)
            self.print1ToMaxOfNDigitsRecursively(number, n, 0)
            
    def printNumber(self, number):
        isBegin = True
        nLength = len(number)
        for i in range(nLength):
            if isBegin and number[i] != "0":
                isBegin = False
            if not isBegin:
                print(number[i])
        print("\t")
        
    
    def print1ToMaxOfNDigitsRecursively(self, number, length, index):
        if index == length-1:
            self.printNumber(number)
            return
        for i in range(10):
            number[index+1] = str(i)
            self.print1ToMaxOfNDigitsRecursively(number, length, index+1)
```

```python
Solution().print1ToMaxOfNDigits(2)
```
